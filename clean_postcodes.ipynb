```python
import pandas as pd 
```

```python
# Import and read CSV for postcodes in Australia
postcodes_df = pd.read_csv('australian_postcodes.csv')
postcodes_df.head() 
```

```python
# Keep only relevant columns for analysis 
relevant_postcodes = postcodes_df[['postcode','long','lat']]

# Keep Postcodes from 6000 to 6200 (WA)
relevant_postcodes = relevant_postcodes[(relevant_postcodes['postcode'] <= 6200) & (relevant_postcodes['postcode'] >=6000)]

# Remove rows with duplicate postcodes
relevant_postcodes = relevant_postcodes.drop_duplicates(subset='postcode')

# # Rename columns to suit origina dataframe 
relevant_postcodes = relevant_postcodes.rename(columns={'postcode' : 'POSTCODE', 
                                                       'long' : 'LONG',
                                                        'lat' : 'LAT'})          
# Check data types. 
relevant_postcodes.dtypes
```

```python
relevant_postcodes.to_csv("clean_postcodes.csv", header=True, index =False)
```
